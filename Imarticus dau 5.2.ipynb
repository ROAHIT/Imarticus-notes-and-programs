```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/0.MKCE/Datasets/HR_comma_sep.csv')
```

```python
df
```

```python
# splitting
import numpy as np

#x = df.drop(['left','Department','salary'],axis=1).values
x = df.select_dtypes(include=np.number)
y = df['left'].values
x.drop(['left'],axis=1)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# logistic regression

from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
```

```python
df.dtypes
```

```python
model.fit(x_train,y_train)
```

```python
# prediction 

y_pred=model.predict(x_test)

y_pred
```

```python
# probability of occurence 0 or 1 in place of corresponding x_test values

model.predict_proba(x_test)
```

```python
x_test
```

```python
# measuring the accuracy of model

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
z=classification_report(y_test,y_pred)
print(z)
```

```python
# filter out the numerical variables and consider only the categorical variables using (include=object)
df_categoric = df.select_dtypes(include = object)

# display categorical features
df_categoric.columns
```

```python
# to create the dummy variables  we use 'get_dummies()' from pandas 
# to create (n-1) dummy variables we use 'drop_first = True' 
dummy = pd.get_dummies(df_categoric, drop_first = True)
```

```python
# concatenate the numerical and dummy encoded categorical variables column-wise
df_dummy = pd.concat([x, dummy], axis=1)

# display data with dummy variables
df_dummy.head()
```

```python
df_dummy
```

```python
# splitting

x = df_dummy.values
y = df['left'].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# logistic regression

from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# prediction 

y_pred=model.predict(x_test)

y_pred
```

```python
# probability of occurence 0 or 1 in place of corresponding x_test values

model.predict_proba(x_test)
```

```python
# measuring the accuracy of model

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
z=classification_report(y_test,y_pred)
print(z)
```
